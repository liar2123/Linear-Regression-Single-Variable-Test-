```python
#These are the libraries we will be using for this lab
import pandas as pd #For data manipulation and analysis
import numpy as np #For numerical operations
import matplotlib.pyplot as plt #For plotting and visualization
from sklearn import linear_model #For linear regression modeling
```

```python
df = pd.read_csv('https://github.com/liar2123/Linear-Regression-Single-Variable-Test-/blob/main/salary_data.csv?raw=true_values=') 
#Read the data from the csv file
df #Display the data to check if it has been read correctly
```

```python
%matplotlib inline
#Plotting the data to visualize the relationship between years of experience and salary
plt.scatter(df['YearsExperience'], df['Salary'], color='blue') #Scatter plot of the data points
plt.xlabel('Years of Experience') #Label for x-axis
plt.ylabel('Salary') #Label for y-axis
plt.title('Salary vs Years of Experience') #Title for the plot
```

```python
reg = linear_model.LinearRegression() #Create an instance of the LinearRegression class
reg.fit(df[['YearsExperience']], df['Salary']) #Fit the linear regression model to the data
#The fit method takes the independent variable (YearsExperience) and the dependent variable (Salary)
```

```python
reg.predict([[5]]) #Predict the salary for 5 years of experience using the trained model
#The predict method always takes x as a 2D array, so we need to pass [[5]] instead of just 5
```

```python
#Manually calculating the predicted salary for 5 years of experience using the coefficients from the trained model
#The formula for linear regression is: y = mx + b, where m is the coefficient and b is the intercept
```

```python
#First, we need to get the coefficient (m)
reg.coef_ #This will give us the coefficient (m) of the linear regression model
```

```python
#Second, we need to get the intercept (b)
reg.intercept_ #This will give us the intercept (b) of the linear regression model
```

```python
#Then we can use the formula to calculate the predicted salary for 5 years of experience
#y = m*x + b
9449.96232146*5 + 25792.20019866871 #This will give us the predicted salary for 5 years of experience using the coefficients from the trained model
```

```python
#Adding plotting the regression line to the scatter plot
%matplotlib inline
#Plotting the data to visualize the relationship between years of experience and salary
plt.scatter(df['YearsExperience'], df['Salary'], color='blue') #Scatter plot of the data points
plt.xlabel('Years of Experience') #Label for x-axis
plt.ylabel('Salary') #Label for y-axis
plt.title('Salary vs Years of Experience') #Title for the plot
plt.plot(df['YearsExperience'], reg.predict(df[['YearsExperience']]), color='red') #Plotting the regression line
```
